```python
from langchain.llms import GooglePalm
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

# Initialize the Google Palm model (replace with your API key)

api_key = ''
llms = GooglePalm(google_api_key=api_key, temperature=0.1)
```

```python
# Define a prompt template for the PaLM model
prompt_template = PromptTemplate(
    template="Evaluate the following resume against the job description and provide an ATS score (out of 100):\n\nResume:\n{resume}\n\nJob Description:\n{description}\n\nScore:"
)


```

```python
# Create a LangChain chain
chain = LLMChain(llm=llms, prompt=prompt_template)
```

```python
import re
import pandas as pd
import concurrent.futures
import time
from ratelimit import limits, sleep_and_retry

# Constants
RATE_LIMIT = 70  # Maximum requests per minute
ONE_MINUTE = 60  # Seconds in a minute

# Function to process each resume and job description pair
@sleep_and_retry
@limits(calls=RATE_LIMIT, period=ONE_MINUTE)
def process_row(index, row):
    resume = row['Resume_str']
    description = row['description']
    
    try:
        # Get the ATS score from the PaLM model
        response = chain.run(resume=resume, description=description)
        
        # Extract the first number from the response
        match = re.search(r'\d+', response)
        if match:
            ats_score = int(match.group())
            # Ensure the ATS score is within the 0-100 range
            if ats_score < 0:
                ats_score = 0
            elif ats_score > 100:
                ats_score = 100
        else:
            ats_score = None  # Handle any unexpected response format
    except Exception as e:
        print(f"Error processing row {index}: {e}")
        ats_score = None
    
    print(f"Row {index} ATS Score: {ats_score}")
    return (index, ats_score)

# Initialize an empty list to store scores
scores = [None] * len(df)

# Process each resume and job description pair using multithreading
with concurrent.futures.ThreadPoolExecutor() as executor:
    futures = {executor.submit(process_row, index, row): index for index, row in df.iterrows()}
    
    for future in concurrent.futures.as_completed(futures):
        index, ats_score = future.result()
        scores[index] = ats_score



```

```python

# Add the 'ATS_Score' column to the existing DataFrame
df['ATS_Score'] = scores

```

```python
# Save the updated DataFrame to a CSV file
df.to_csv(r'E:\Class\sem3\nlp\Project\final_ats_scores1.csv', index=False)
```
